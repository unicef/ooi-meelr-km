POC of llamaindex with milvus as vector store

```python
import logging
import sys

# Uncomment to see debug logs
logging.basicConfig(stream=sys.stdout, level=logging.DEBUG)
logging.getLogger().addHandler(logging.StreamHandler(stream=sys.stdout))

from llama_index.core import VectorStoreIndex, SimpleDirectoryReader, Document
from llama_index.vector_stores.milvus import MilvusVectorStore
from IPython.display import Markdown, display
import textwrap

# for creating index over documents
from llama_index.core import StorageContext
import os
```

```python
# load documents
documents = SimpleDirectoryReader("/proj/staff/jayant/rag/ooi-meelr-km/data").load_data()

print("Document ID:", documents[0].doc_id)
```

```python
vector_store = MilvusVectorStore(
    dim=1536,
    overwrite=True,
    enable_sparse=True,
    hybrid_ranker="RRFRanker",
    hybrid_ranker_params={"k": 60},
    #BUG: figure out uri for milvus
    # uri="/proj/staff/jayant/.milvus_llamaindex.db" # uri="./milvus_demo.db",
)
storage_context = StorageContext.from_defaults(vector_store=vector_store)
index = VectorStoreIndex.from_documents(
    documents, storage_context=storage_context
)
```
